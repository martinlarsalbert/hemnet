```python
import geopandas as gpd
import pandas as pd
import folium
from folium.plugins import HeatMap

#house_data = pd.read_csv('house_data.csv')
house_data = pd.read_csv('house_data_clean.csv')

#feature = 'Pris per kvadratmeter'
feature = 'price'
mask = house_data[feature].notnull()
house_data = house_data.loc[mask]



max_amount = float(house_data[feature].max())

hmap = folium.Map(location=[57.7, 12], zoom_start=12, )

hm_wide = HeatMap( list(zip(house_data.lat.values, house_data.lon.values, house_data[feature].values)),
                   min_opacity=0.0001,
                   max_val=max_amount,
                   radius=15, blur=30, 
                   max_zoom=15, 
                 )

#folium.GeoJson(district23).add_to(hmap)
hmap.add_child(hm_wide)
```
